```python
import pandas as pd
import os
```

```python
ne_path = "./data/korea/NIKL_NE_2022_CSV_개체명 분석"
el_path = "./data/korea/NIKL_EL_2022_CSV_개체 연결"
dx_path = "./data/korea/XXDX2200000001.csv"
```

```python
def read_csv(path):
    if os.path.isdir(path):
        file_list = [file for file in os.listdir(path) if file.endswith(".csv")]
        result_df = pd.DataFrame()
        for file in file_list:
            df = pd.read_csv(os.path.join(path, file), index_col=0)
            result_df = pd.concat([result_df, df])
    else:
        result_df = pd.read_csv(path, index_col=0)
    return result_df
```

```python
ne_df = read_csv(ne_path)
ne_df.head()
```

```python
el_df = read_csv(el_path)
el_df.head()
```

```python
dx_df = read_csv(dx_path)
dx_df.head()
```
